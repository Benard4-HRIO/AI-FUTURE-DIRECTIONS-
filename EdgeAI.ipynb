```python
import tensorflow as tf
import tensorflow_datasets as tfds

# Load dataset
dataset, info = tfds.load('rock_paper_scissors', with_info=True, as_supervised=True)
train_ds, val_ds = dataset['train'], dataset['test']

# Normalize and resize images to 100x100
def preprocess(image, label):
    image = tf.image.resize(image, [100,100])
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

train_ds = train_ds.map(preprocess).batch(32).shuffle(1000)
val_ds = val_ds.map(preprocess).batch(32)

# Check number of classes
num_classes = info.features['label'].num_classes
print("Number of classes:", num_classes)
```

```python
from tensorflow.keras import layers, models

# Build a small CNN
model = models.Sequential([
    layers.Conv2D(16, (3,3), activation='relu', input_shape=(100,100,3)),  # images are 100x100
    layers.MaxPooling2D(2,2),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes, activation='softmax')  # 3 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Show model summary
model.summary()
```

```python
# Train the model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10  # you can change to 5 if training is slow
)
```

```python
# Train the model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10  # or 5 if you want faster training
)
```

```python
# Evaluate on validation set
loss, accuracy = model.evaluate(val_ds)
print("Validation Accuracy:", accuracy)
print("Validation Loss:", loss)
```

```python
# Convert to TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save TFLite model
with open('rps_classifier.tflite', 'wb') as f:
    f.write(tflite_model)

print("TFLite model created successfully! Size:", len(tflite_model)/1024, "KB")
```

```python
import numpy as np
from PIL import Image
import tensorflow as tf

# Load TFLite model
interpreter = tf.lite.Interpreter(model_path="rps_classifier.tflite")
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Function to classify one image
def classify_image(path):
    img = Image.open(path).resize((100,100))
    img = np.array(img)/255.0
    img = np.expand_dims(img, axis=0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], img)
    interpreter.invoke()
    pred = interpreter.get_tensor(output_details[0]['index'])
    return pred

# Example usage (replace with your own image)
# result = classify_image("rock_image.jpg")
# print("Predictions:", result)
```
